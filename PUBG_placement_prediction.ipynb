## Simple Assumptions
Before looking into the data, we can make some assumptions based on our prior understanding of the game and features.

Features that have correlation with the players ability.
- **assists, damageDealt, DBNOs, headshotKills, killPlace, killPoints, kills, killStreaks, longestKill**: These are features that describe the amount of damage the player dealt during the one match. We can assume that the higher these values are, the player is high-skilled, or at least that the player survived longer in that particular match.
- **revives**: This is a features that is related to team cooperation. It might have positive correlation with winning.

Features that must have higher values if the player survives longer
- **rideDistance, swimDistance, walkDistance**: These features measure the amount of movement of the player. They must show positive correlation with the time of survival of the player.

Features that can be combined
- **boosts, heals**: Although boosting and healing items are different, they serve the same purpose in maintaining player health. Therefore, the can be combined into one feature.

Features that fo not need to be taken into consideration for model training
- **Id, goupId**: These are features that merely describe the identity of the player, and therefore the values will show no correlation with winning.
- **matchId, matchDuration**: These are features that merely describe the match, and therefore the values will have no correlation with winning.
- **rankPoints, winPoints, killPoints**: These feature have too many missing values.

# 3. Data Wrangling
## Wrangling Hypothesis

Before cleaning the dataset, we can make think of several methods of cleaning.
#### 1. Anomaly Detection & Removal

There are players that hack the game system. We can assume that the data of these cheaters are anomalies, and therefore deleting them might result in better prediction.
#### 2. Data Normalization

Because each match has a different number of players joined, we might need to normalize features such as kills or damage. This is because killing 2 players out of a 100 should be weighted differently than killing 2 players out of 10.
#### 3. Feature Generation

Features that have similar characteristics, such as heals and boosts or features related to movement, can be combined to create a new features with better prediction power. Also, features can be created to represent new information, such as the number of players in each group. New features may help training the model.
#### 4. Feature Removal

Because simple models tend to be better, we can remove features that are highly correlated to other existing features. Also, selecting only important features can creat a more simple model.
#### 5. Data Deletion

The dataset includes data that we do not want to learn, such as event mode data. Removing these will help us focus on data we wish to predict
#### 6. Grouping

PUBG is a team game in which the final placement is based on teams, not individual players. Therefore, the best placement among team members will be more important than others. We can train our model based on groups, not individuals.

As we have several methods of wrangling, we can create several dataframes that may or may not wrangle the data accordingly.
